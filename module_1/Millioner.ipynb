# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
directorsDict = defaultdict(list)

for _,row in data.iterrows():
    imdb_id = row['imdb_id']
    for dir in row['director'].split('|'):
        directorsDict['imdb_id'].append(imdb_id)
        directorsDict['director'].append(dir)

directorDF = pd.DataFrame(directorsDict)

dd = directorDF.merge(data,on='imdb_id',how='inner')

total = dd.groupby(['director_x'])['original_title'].agg(['count'])
td = total.reset_index()

profit = dd[dd['total_income'] > 0].groupby(['director_x'])['original_title'].agg(['count'])
pd = profit.reset_index()

joined = pd.merge(td, on='director_x', how='inner')
joined.columns = ['director','profitfilms','totalfilms']
joined['percentage'] = joined['profitfilms']/joined['totalfilms']

display(joined[joined.percentage == joined.percentage.max()].sort_values(["totalfilms"],ascending=False).head(20))

answer_ls.append(4)
```
